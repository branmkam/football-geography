```python
import pandas as pd
import numpy as np
```

```python
#import
clubs = pd.read_csv('clubs.csv')
logos = pd.read_csv('withlogos.csv')
logos = logos[['clubLabel', 'logoImage']]
logos
```

```python
#drop duplicate stadiums
clubs = clubs.drop_duplicates(subset=['clubLabel'])
clubs
```

```python
clubs = pd.merge(left = clubs, right = logos, how = 'outer', on = 'clubLabel' )
```

```python
uClubs = {}
for c in clubs.columns:
    uClubs[c] = c[:-5] if c[-5:] == 'Label' else c
clubs = clubs.rename(columns=uClubs)
```

```python
clubs['coordinates'] = [x[6:-1] for x in clubs['coordinates']]
for c in clubs.columns[:-1]:
    clubs[c] = [x.replace('"', '') for x in clubs[c]]
```

```python
#United Kingdom Issue
clubs.loc[clubs['league'].str.find('Scottish') != -1, 'country'] = 'Scotland'
clubs.loc[clubs['league'].str.find('Highland') != -1, 'country'] = 'Scotland'
clubs.loc[clubs['league'].str.find('NIFL') != -1, 'country'] = 'Northern Ireland'
clubs.loc[clubs['league'].str.find('Welsh') != -1, 'country'] = 'Wales'
clubs.loc[clubs['league'].str.find('Cymru') != -1, 'country'] = 'Wales'
clubs.loc[clubs['league'].str.find('Gwent') != -1, 'country'] = 'Wales'
clubs.loc[clubs['country'].str.find('United Kingdom') != -1, 'country'] = 'England'
clubs.loc[clubs['league'].str.find('Gibraltar') != -1, 'country'] = 'United Kingdom'
clubs.loc[clubs['league'].str.find('Priaulx') != -1, 'country'] = 'United Kingdom'
clubs.loc[clubs['league'].str.find('Cayman') != -1, 'country'] = 'United Kingdom'
```

```python
clubs.to_csv('clubs_official.csv')
jsonfinal = clubs.to_json().replace('\\/', '/')
with open('clubs_official.json', "w") as f:
    f.write(jsonfinal)
```
